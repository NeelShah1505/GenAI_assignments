```python
# 1. Install the library
!pip install gTTS

from gtts import gTTS
import IPython.display as ipd
import os

# 2. Define the text to be converted
# You can change this to any sentence you like!
text_input = "I consider myself one of the passionate few."

# 3. Initialize the TTS engine
# lang='en' for English; slow=False for normal speed
tts = gTTS(text=text_input, lang='en', slow=False)

# 4. Save the output to an MP3 file
audio_file = "assignment8_output.mp3"
tts.save(audio_file)

print(f"--- Audio Generation Complete ---")
print(f"File saved as: {audio_file}")

# 5. Play the audio directly in your Colab notebook
ipd.Audio(audio_file)
```
