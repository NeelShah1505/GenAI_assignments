```python
# 1. Install required libraries
!pip install diffusers transformers accelerate

import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# 2. Load the pre-trained Stable Diffusion model
# We use torch.float16 to save memory and speed up generation on the T4 GPU
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# 3. Define your creative prompt
prompt = "A majestic palace in Pune at sunset, digital art, high resolution, cinematic lighting"

# 4. Generate the image
# num_inference_steps: higher means more detail (usually 30-50)
image = pipe(prompt, num_inference_steps=50).images[0]

# 5. Display and save
plt.imshow(image)
plt.axis("off")
plt.title(f"Generated Art: {prompt}")
plt.show()

image.save("assignment4_artwork.png")
```
