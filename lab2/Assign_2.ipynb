```python
from google import genai
from google.genai import types

client = genai.Client(api_key=API_KEY)

def run_lab_experiment(technique_name, user_prompt, system_instruction="You are a helpful GenAI Lab Assistant."):
    print(f"\n{'='*20} {technique_name} {'='*20}")

    try:
        # Use the updated 'gemini-2.5-flash' model
        response = client.models.generate_content(
            model='gemini-2.5-flash',
            config=types.GenerateContentConfig(
                system_instruction=system_instruction,
                temperature=0.7
            ),
            contents=user_prompt
        )
        print(response.text)
    except Exception as e:
        print(f"Error: {e}")
    print(f"{'='*60}\n")

# --- PROMPTS ---

# A. INTERVIEW APPROACH
interview_query = "I want to design a solar backup for my office in Pune. Ask me exactly 3 questions one-by-one to understand my needs."

# B. CHAIN OF THOUGHT (CoT)
cot_query = "A solar array has 20 panels. 4 are 250W, 10 are 300W, and 6 are broken. Total kW output? Think step-by-step."

# C. TREE OF THOUGHTS (ToT)
tot_query = """Task: Strategy to recycle solar batteries in India.
Expert 1 (Env), Expert 2 (Econ), and Expert 3 (Logistics) discuss.
Provide a 3-step consensus strategy."""

# D. ZERO-SHOT
zero_shot_query = "Classify: 'The inverter is noisy but efficient.' Output: [Positive/Negative/Neutral]"

# E. FEW-SHOT
few_shot_query = """
Review: 'Fast shipping!' -> Sentiment: Positive
Review: 'High price, low quality.' -> Sentiment: Negative
Review: 'The inverter is noisy but efficient.' -> Sentiment:"""

# --- EXECUTION ---
run_lab_experiment("1. INTERVIEW APPROACH", interview_query)
run_lab_experiment("2. CHAIN OF THOUGHT (CoT)", cot_query)
run_lab_experiment("3. TREE OF THOUGHTS (ToT)", tot_query)
run_lab_experiment("4. ZERO-SHOT PROMPTING", zero_shot_query)
run_lab_experiment("5. FEW-SHOT PROMPTING", few_shot_query)
```
