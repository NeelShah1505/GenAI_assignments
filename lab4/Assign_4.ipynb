```python
from google import genai
from google.colab import userdata


GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
client = genai.Client(api_key=GOOGLE_API_KEY)

def call_gemini(prompt):

    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=[
            {"role": "user", "parts": [{"text": "Act as a Hospital receptionist of Pune Hospital. You have all kinds of tests available for the patient to do. if a patient asks for a test for thier problem suggest them the best 3 he/she can go for right now"}]},
            {"role": "model", "parts": [{"text": "Certainly! I'm here to help. What seems to be the problem, or what kind of tests are you interested in today?"}]},
            {"role": "user", "parts": [{"text": prompt}]}
        ]
    )
    print(response.text)

# Testign
call_gemini("I have been having severe chest pain and shortness of breath.")
```
